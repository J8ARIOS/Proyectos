```python
#RANDOM PASSWORD GENERATOR

import random
import string

respuesta= "s"

while respuesta == "s":   

    Numero_Caracteres= input("¿Cuantos caracteres quieres que tenga la contraseña?")
    Caracteristicas= input("¿Que caracteristicas deseas?: 1)Solo minusculas.     2)Solo mayusculas.  3)Solo numeros.     4)Solo simbolos.    5)Minusculas con mayusculas.    6)Minusculas con numeros.   7)Minusculas con simbolos.  8)Mayusculas con numeros.   9)Mayusculas con simbolos.  10)Numeros con simbolos. 11)Todo combinado")
    Impresiones= 0
    
    
    if Numero_Caracteres.isdigit():
        print

    else:
        print("El caraacter seleccionado es invalido")
        break 
 

    if Caracteristicas == str(1):
        while(Impresiones < int(Numero_Caracteres)):
            print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
            random_minuscula=  random.choice(string.ascii_lowercase)
            print(random_minuscula, end="")
            Impresiones = Impresiones + 1
    elif Caracteristicas == str(2):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            random_mayuscula = random.choice(string.ascii_uppercase)
            print(random_mayuscula, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(3):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            random_number= (random.randint(0, 9))
            print(random_number,end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(4):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            simbolos= "!@#$%^&*()_+-=[]{}|;:,.<>/?"
            random_symbol= random.choice(simbolos)
            print(random_symbol, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(5):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            random_letter=  random.choice(string.ascii_letters)
            print(random_letter, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(6):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            caracter_aleatorio = random.choice(string.digits + string.ascii_lowercase)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(7):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            simbolos= "!@#$%^&*()_+-=[]{}|;:,.<>/?"
            caracter_aleatorio = random.choice(string.ascii_lowercase + simbolos)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(8):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            caracter_aleatorio = random.choice(string.ascii_uppercase + string.digits)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(9):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            simbolos= "!@#$%^&*()_+-=[]{}|;:,.<>/?"
            caracter_aleatorio = random.choice(string.ascii_uppercase + simbolos)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(10):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            simbolos= "!@#$%^&*()_+-=[]{}|;:,.<>/?"
            caracter_aleatorio = random.choice(string.digits + simbolos)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1

    elif Caracteristicas == str(11):
        print("Esta es la contraseña generada aleatoriamente segun las caracteriticas seleccionadas:")
        while(Impresiones < int(Numero_Caracteres)):
            simbolos= "!@#$%^&*()_+-=[]{}|;:,.<>/?"
            caracter_aleatorio = random.choice(string.ascii_letters + string.digits + simbolos)
            print(caracter_aleatorio, end="")
            Impresiones = Impresiones + 1
    else:
        print("La caracteristica seleccionada no existe")

    print("")
    print("----------------------------------------")


    respuesta= input("¿Desea volver a ejecutar el programa? (s/n)")
    if respuesta == "s":
        continue
    elif respuesta == "n":
        break
    else:
        print("La opcion elegida no exixte")

  
     
    
```
