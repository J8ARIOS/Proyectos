```python
while True: 

    materia = input("¿De que materia estas calculando la nota definitiva?")

    #COGNITIVO
    ficho = False
    while ficho == False:
        num_notas = input("¿Cuantas notas son en cognitivo?")
        if num_notas.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado es invalido")

    notas_acumuladas = []
    
    impresiones = 1
    while int(num_notas) >= impresiones:
        ficho = False
        while ficho == False:
            notas = input("Dame la nota #" + str(impresiones))
            if notas.isalpha():
                print("ERROR: El valor proporcionado es invalido")
            else:
                ficho = True
                notas_acumuladas.append(float(notas))
                impresiones = impresiones + 1
                #print(notas_acumuladas) #control

    ficho = False
    while ficho == False:
        porcentaje = input("¿Que porcentaje tiene el cognitivo en la materia?")
        if porcentaje.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado no es valido")

    porcentaje = int(porcentaje) / 100

    total = 0
    for valor in notas_acumuladas:
        total += valor
    #print(total) #control

    nota_sin = total / int(num_notas)
    #print(nota_sin)#control
    nota_final1 = nota_sin * porcentaje
    #print(nota_final1) #control


    #PROCEDIMENTAL
    ficho = False
    while ficho == False:
        num_notas = input("¿Cuantas notas son en procedimental?")
        if num_notas.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado es invalido")

    notas_acumuladas = []
    
    impresiones = 1
    while int(num_notas) >= impresiones:
        ficho = False
        while ficho == False:
            notas = input("Dame la nota #" + str(impresiones))
            if notas.isalpha():
                print("ERROR: El valor proporcionado es invalido")
            else:
                ficho = True
                notas_acumuladas.append(float(notas))
                impresiones = impresiones + 1
                #print(notas_acumuladas) #control

    ficho = False
    while ficho == False:
        porcentaje = input("¿Que porcentaje tiene el procedimental en la materia?")
        if porcentaje.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado no es valido")

    porcentaje = int(porcentaje) / 100

    total = 0
    for valor in notas_acumuladas:
        total += valor
    #print(total) #control

    nota_sin = total / int(num_notas)
    #print(nota_sin) #control
    nota_final2 = nota_sin * porcentaje
    #print(nota_final2) #control


    #ACTITUDINAL
    ficho = False
    while ficho == False:
        num_notas = input("¿Cuantas notas son en actitudinal?")
        if num_notas.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado es invalido")

    notas_acumuladas = []
    
    impresiones = 1
    while int(num_notas) >= impresiones:
        ficho = False
        while ficho == False:
            notas = input("Dame la nota #" + str(impresiones))
            if notas.isalpha():
                print("ERROR: El valor proporcionado es invalido")
            else:
                ficho = True
                notas_acumuladas.append(float(notas))
                impresiones = impresiones + 1
                #print(notas_acumuladas) #control

    ficho = False
    while ficho == False:
        porcentaje = input("¿Que porcentaje tiene el actitudinal en la materia?")
        if porcentaje.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado no es valido")

    porcentaje = int(porcentaje) / 100

    total = 0
    for valor in notas_acumuladas:
        total += valor
    #print(total) #control

    nota_sin = total / int(num_notas)
    #print(nota_sin) #control
    nota_final3 = nota_sin * porcentaje
    #print(nota_final3) #control


    #BIMENSUAL
    ficho = False
    while ficho == False:
        bimensual = input("Resultado en el bimensual")
        if bimensual.isdigit():
            ficho = True
        else:
            print("ERROR: Valor proporcionado invalido")

    ficho = False    
    while ficho == False:
        porcentaje = input("¿Que porcentaje tiene el bimensual en la materia?")
        if porcentaje.isdigit():
            ficho = True
        else:
            print("ERROR: El valor proporcionado no es valido")

    porcentaje = int(porcentaje) / 100
    bimensual = float(bimensual) * porcentaje
    #print(bimensual) #control

    nota_definitiva = nota_final1 + nota_final2 + nota_final3 + bimensual
    print("\nEsta en la nota definitiva en " + materia + "\n --->" ,nota_definitiva )

    opcion = input("¿Quieres volver a ejecutar el programa? (s/n)")
    if opcion == "n":
        break
    elif opcion == "s":
        print("--------------------------------")
        continue
    else:
        print("Opcion invalida")
        break
    #prueba commit
```
